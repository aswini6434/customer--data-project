```python
%pip install gender_guesser
```

```python
import gender_guesser.detector as Gender

```

```python
detector = Gender.Detector()
```

```python

df["Gender_predicted"] = df["FirstName"].apply(lambda x:detector.get_gender(x) if pd.notnull(x) else None)

df["Gender_predicted"] = df["Gender_predicted"].replace({
    "male": "M",
    "mostly_male": "M",
    "female": "F",
    "mostly_female": "F",
    "andy": "unknown",
    "unknown": "unknown"
})

df["Gender"] = df["Gender"].fillna(df["Gender_predicted"])
```

```python
df["Age"] = df["Age"].replace("",np.nan).astype(float)
df["Age"] = df["Age"].fillna(df["Age"].mean()).astype(int)
```

```python
df["Age"]
```

```python
df['Age'].isnull().sum()
```

```python
df['City'] = df['City'].fillna('Unknown')

```

```python
df["Email"]
```

```python
df['Email'] = df['Email'].str.replace(r'\.(biz|info|org)$', '.com', regex=True)

df['Email'] = df['Email'].fillna(df['FirstName'].str.lower() + '.' + df['LastName'].str.lower() + '@example.com')

```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#spending vs orders 
sns.scatterplot(data=df, x='NumberOfOrders', y='TotalSpend', hue='Gender_predicted')
```

```python
#customer registration over time 

df['RegisteredDate'] = pd.to_datetime(df['RegisteredDate'])
df['RegisteredMonth'] = df['RegisteredDate'].dt.to_period('M')

df.groupby('RegisteredMonth').size().plot(kind='bar')
```

```python
#customer count by city
df['City'].value_counts().head(10).plot(kind='bar')
```

```python
# Top customers by spend 
top_customers = df.sort_values('TotalSpend', ascending=False).head(20)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_customers, x='CustomerID', y='TotalSpend', palette='viridis')

plt.title('Top 30 Customers by Total Spend')
plt.xlabel('Customer ID')
plt.ylabel('Total Spend')
plt.xticks(rotation=35)
plt.tight_layout()
plt.show()
```

```python
# gender distribution 
df['Gender_predicted'].value_counts().plot(kind='pie', autopct='%1.1f%%',explode=[0, 0.1])
```

```python
#All customers 

plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='CustomerID', y='TotalSpend', color='teal')

plt.title('Customer Spend Distribution')
plt.xlabel('Customer ID')
plt.ylabel('Total Spend')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# customers registration 
df['RegisteredDate'] = pd.to_datetime(df['RegisteredDate'], errors='coerce')

df['RegisteredMonth'] = df['RegisteredDate'].dt.to_period('M')
monthly_counts = df['RegisteredMonth'].value_counts().sort_index()



plt.figure(figsize=(12, 5))
sns.barplot(x=monthly_counts.index.astype(str), y=monthly_counts.values, color='teal')
plt.title('Customer Registrations by Month')
plt.xlabel('Month')
plt.ylabel('Number of Registrations')
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```
